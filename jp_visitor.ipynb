```python
import pdfplumber
import pandas as pd
import re
```

```python
pdf = pdfplumber.open("data/Visitor_Japanese_Overseas_Travelers.pdf")
```

```python
page_jpn_issue = pdf.pages[0]
```

```python
jpn_visitor = page_jpn_issue.extract_table()
df_jp_visitor = pd.DataFrame(jpn_visitor[3:], columns=jpn_visitor[1])
```

```python
df_jp_visitor.rename(columns={'訪日外客数 (＊1)\nVisitor\nArrivals': 'Visitor Arrivals', '出国日本人数 (＊2)\nJapanese Overseas\nTravelers': 'Japanese Overseas Travelers'}, inplace=True)
df_jp_visitor
```

```python
df_jp_visitor.to_csv("jp_visitor.csv", index=False)
```
